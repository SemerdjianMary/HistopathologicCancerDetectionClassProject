```python
maxC = 10000
```

```python
# Load the first 10000 RGB TIFF images and convert them to arrays
histo_img = []

for i, img_path in enumerate(os.listdir(img_dir)):
    if i >= maxC:
        break
    img = load_img(os.path.join(img_dir, img_path), target_size=(img_width, img_height))
    img_array = img_to_array(img)
    histo_img.append(img_array)
   
```

```python
histo_img = np.array(histo_img)
```

```python
histo_img = histo_img.flatten()
```

```python
# Normalize the input data
histo_img = histo_img / 255.
```

```python
# Creating pandas dataframe using the flattened images list
X = pd.DataFrame(scale(histo_img))
print(X.shape)
X.head()
```
